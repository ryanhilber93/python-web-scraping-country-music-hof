```python
name_all_list = []
date_all_list = []
date_all_list2 = []
for i in range(1, 17) :
    website_url_all = 'https://countrymusichalloffame.org/hall-of-fame/members/page/' + str(i)
    response_all = requests.get(website_url_all)
    hof_all = BeautifulSoup(response_all.content, 'lxml')
    for artist in hof_all.find_all('h3'):
        name_all_list.append(artist.text)
    for artist in hof_all.find_all('div', attrs = {'class': 'vertical-card_content--copy'}):
        date_all_list.append(artist.text)
for element in date_all_list :
    new_string = element.replace('Inducted','').strip()
    date_all_list2.append(new_string)
```

```python
name_all_list
```

```python
date_all_list2
```

```python
df_all = pd.DataFrame()
df_all['name'] = name_all_list
df_all['date_inducted'] = date_all_list2
df_all
```

```python
inductions_by_year = df_all['date_inducted'].value_counts()
```

```python
inductions_by_year
```

```python
inductions_by_year2 = inductions_by_year.to_frame().reset_index()
```

```python
inductions_by_year2
```

```python
inductions_by_year3 = inductions_by_year2.rename(columns={"index": "year", "date_inducted": "number_inducted"})
inductions_by_year3
```

```python
inductions_by_year4 = inductions_by_year3.sort_values('year')
inductions_by_year4
```

```python
inductions_by_year4.plot.bar(x='year', y='number_inducted', figsize=(20,10), legend=None)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Inductees', fontsize=15)
plt.text(30,12.7, 'Ten "Special Inductees" in 2001', fontsize='large')
plt.arrow(34,12.5,3,-1,width=.1,ec='green')
plt.savefig('inductees_by_year.png');
```
